hate_speech_score - continuous hate speech measure, where higher = more hateful and lower = less hateful. > 0.5 is approximately hate speech, < -1 is counter or supportive speech, and -1 to +0.5 is neutral or ambiguous.

Doing bucketing of labels

```python
def classify_hate_speech(score):
    if score > 0:
        return 1  # Hate speech
    else:
        return 0  # Supportive speech
```

```python
df['hate_speech_score'] = df['hate_speech_score'].apply(classify_hate_speech)
```

```python
df.head()
```

```python
df = df.rename(columns={'hate_speech_score': 'label'})
```

#Sampling the datasets

```python
%cd /content/drive/MyDrive/ANLP_indiv_project
```

```python
import pandas as pd

def load_datasets():
    datasets = {
        "DAVIDSON": "datasets/processed_DAVIDSON.csv",
        "DYNABENCH": "datasets/processed_DYNABENCH.csv",
        "MHS": "datasets/processed_MHS.csv",
        "MLMA": "datasets/processed_MLMA.csv"
    }

    sampled_datasets = {}

    for name, path in datasets.items():
        df = pd.read_csv(path)
        sampled_df = df.sample(frac=0.01, random_state=42)  # Sample 1% of the data
        sampled_datasets[name] = sampled_df

    return sampled_datasets

# Load the sampled datasets
sampled_datasets = load_datasets()

# Save the sampled datasets to new CSV files if needed
for name, df in sampled_datasets.items():
    df.to_csv(f"datasets/sampled_reduced_{name}.csv", index=False)
```
